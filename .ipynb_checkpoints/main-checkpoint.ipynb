## Filtering for data on Singapore

```python
mask=suicides.country == "Singapore"
suicides_singapore=suicides[mask]
suicides_singapore.head(20)
```

```python
suicides_singapore.groupby("year").suicides_no.sum()
suicides_singapore.groupby("year").suicides_no.sum().plot(kind="line", color="green", title="Suicides in Singapore, Males and Females, 1979-2015")
```

```python
suicides_singapore[suicides_singapore['sex'] == "male"].groupby("year").suicides_no.sum().plot(kind="line", label="Male", title="Suicides in Singapore, Males vs Females, 1979-2015")
suicides_singapore[suicides_singapore['sex'] == "female"].groupby("year").suicides_no.sum().plot(kind="line", color="pink",label="Female")
plt.show()
```

```python
suicides_singapore[suicides_singapore['age'] == "75+ years"].groupby("year").suicides_no.sum().plot(kind="line", color="green", title="Suicides in Singapore, Males and Females, 75+ years")
```

```python
suicides_sgmales_elderly=pd.DataFrame(suicides_singapore[(suicides_singapore.sex == 'male') & (suicides_singapore.age == "75+ years")])
suicides_sgmales_elderly.groupby("year").suicides_no.sum().plot(kind="line", title="Suicides in Singapore, Males vs Females, 75+ years")
suicides_sgfemales_elderly=pd.DataFrame(suicides_singapore[(suicides_singapore.sex == 'female') & (suicides_singapore.age == "75+ years")])
suicides_sgfemales_elderly.groupby("year").suicides_no.sum().plot(kind="line", color="pink")
plt.show()
```

```python
suicides_singapore[suicides_singapore['age'] == "55-74 years"].groupby("year").suicides_no.sum().plot(kind="line", color="green",                                                                                                       title="Suicides in Singapore, Males and Females, 55-74 years")
```

```python
suicides_sgmales_seniors=pd.DataFrame(suicides_singapore[(suicides_singapore.sex == 'male') & (suicides_singapore.age == "55-74 years")])
suicides_sgmales_seniors.groupby("year").suicides_no.sum().plot(kind="line", title="Suicides in Singapore, Males vs Females, 55-74 years")
suicides_sgfemales_seniors=pd.DataFrame(suicides_singapore[(suicides_singapore.sex == 'female') & (suicides_singapore.age == "55-74 years")])
suicides_sgfemales_seniors.groupby("year").suicides_no.sum().plot(kind="line", color="pink")
```

```python
suicides_singapore[suicides_singapore['age'] == "35-54 years"].groupby("year").suicides_no.sum().plot(kind="line",  color = "green", title="Suicides in Singapore, Males and Females, 35-54 years")
```
